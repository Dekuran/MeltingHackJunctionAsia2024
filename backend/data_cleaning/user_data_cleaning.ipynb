```python
import pandas as pd
import datetime

AVG_PREGNANCY_DURATION_DAYS = 268 # source = https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3777570/ #TODO fetch from constants - use __init__
```

```python
dummy_df = pd.read_csv("../data/mommytalkDummyData.csv")

display(dummy_df)
```

```python
user_data_df = dummy_df.sort_values('registrationDate')

# add simple id based on registration date order (note would break for multiples of same, not a good id)

user_data_df["userNo"] = range(len(user_data_df))

user_data_df["userId"] = "id_" + user_data_df["userNo"].astype('str')

user_data_df["foodIntolerance"] = "None"

display(user_data_df)

```

```python
# enrich data with useful calculations

user_data_df["childBirth"] = user_data_df["childBirth"].str.replace('04-31','04-30') # spot replace erroneous data - April has 30 days

user_data_df["estConceptionDate"] = pd.to_datetime(user_data_df["dueDate"]) - pd.Timedelta(days=AVG_PREGNANCY_DURATION_DAYS) 

user_data_df["estConceptionOrBirthDate"] = user_data_df["childBirth"].fillna(user_data_df["estConceptionDate"])

user_data_df["pregnancyOrPostBirthWeeks"] = ((pd.to_datetime(datetime.datetime.today())  - pd.to_datetime(user_data_df['estConceptionOrBirthDate'], format='mixed')).dt.days/7).round().astype('int')
```

```python
user_data_df.to_csv("../data/userData.csv")
```
